```python
import pandas as pd
from sklearn.datasets import load_digits
digits= load_digits()
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range(2):
    plt.matshow(digits.images[i])
```

```python
digits.data[:2]
```

```python
df= pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
df['target']=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.2,random_state=2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python

model = SVC(C=2)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.svm import SVC 
model = SVC(gamma=1)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
