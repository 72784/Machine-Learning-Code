```python
import pandas as pd 
import numpy as np
from scipy import stats
```

```python
df=pd.read_csv('heart.csv')
df
```

```python
z_scores = np.abs(stats.zscore(df['Age']))
z_scores

print(np.mean(df['Age']))
print(np.std(df['Age']))

threshold = 3
df_no_outliers = df[(z_scores < threshold)]
df_no_outliers
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])

df['ChestPainType'] = encoder.fit_transform(df['ChestPainType'])
df['RestingECG'] = encoder.fit_transform(df['RestingECG'])
df['ExerciseAngina'] = encoder.fit_transform(df['ExerciseAngina'])
df['ST_Slope'] = encoder.fit_transform(df['ST_Slope'])

df
```

```python
X = df
y = df.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state=30)
```

```python
#Logictic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression() 
model.fit(X_train, y_train) 
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
#SVM
from sklearn import svm
from sklearn import metrics

clf = svm.SVC(kernel='linear') 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier=RandomForestClassifier(n_estimators=100)

classifier.fit(X_train,y_train)
```

```python
Pred = classifier.predict(X_test)
print(Pred) 
```

```python
from sklearn.metrics import classification_report
rand_score=classifier.score(X_test, y_test)
classification_report_rf=classification_report(y_test,Pred)
print("Accuracy score:",rand_score)  
print(classification_report_rf)
```
